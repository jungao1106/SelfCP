```python
import json
xsum = json.load(open('/data1/gj/data/xsum/val.json'))
cicero = json.load(open('/data1/gj/data/declare-lab/cicero/val.json'))
```

```python
res = []
for i in range(1000, 2000):
    d = xsum[i]
    d['task'] = 'sum'
    res.append(d)

for i in range(1000):
    d = cicero[i]
    d['task'] = 'qa'
    res.append(d)   
```

```python
cicero[0]
```

```python
gloden = []
for x in res:
    if x['task'] == 'qa':
        gloden.append(x['Choices'][x['Human Written Answer'][0]])
    else:
        gloden.append(x['summary'])   
```

```python
card0 = []
card1 = []
card2 = []
card3 = []
for i in range(0, 2000, 4):
    card0.append(gloden[i])
    card1.append(gloden[i+1])
    card2.append(gloden[i+2])
    card3.append(gloden[i+3])
```

```python
card = card0 + card1  +card2 + card3
```

```python
fh = open('/data1/gj/PromptCompression/golden/select-ratio.txt', 'w')
for c in card:
    fh.write(c.replace('\n', '') + '\n')
```

```python
fh.close()
```
